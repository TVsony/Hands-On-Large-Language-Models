### Downloading and Running An LLM
The first step is to load our model onto the GPU for faster inference. Note that we load the model and tokenizer separately and keep them as such so that we can explore them separately.

```python
# 2. Import Necessary Modules

# Purpose: Import the modules to load and use a causal language model and tokenizer from Hugging Face.

from transformers import AutoModelForCausalLM, AutoTokenizer

# 3. Load the Model and Tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",  # Text generation model you can use image generation models microsoft/Phi-3.5-vision-instruct
    device_map ="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```


Load the pre-trained causal language model Phi-3-mini-4k-instruct from the Hugging Face repository.

**Parameters:**
**device_map="cuda":** Use GPU for faster computations.

**torch_dtype="auto":** Automatically determines the precision (e.g., float16).

**trust_remote_code=True:** Allows loading custom code from the model's
repository.

**tokenizer**
Load the tokenizer for the same model to convert text into tokens (numerical representations) and back.

```python
# 4. Define the Input Prompt (user input / query )

prompt = "Write an email apologizing to Sarah for the tragic gardening mishap. Explain how it happened.<|assistant|>"

# Purpose: Define the input text that the model will use as context for generating output.
```

```python
# Tokenize the input prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
```

**Purpose: Convert the input prompt (prompt) into token IDs (a tensor format) using the tokenizer.**

**return_tensors="pt":** Return the tokens in PyTorch tensor format.

**.to("cuda"):** Move the tensor to the GPU for faster processing.

```python
# 7. Generate the text from the model
generation_output = model.generate(
  input_ids=input_ids,
  max_new_tokens=20
)

# Purpose: Generate text based on the input prompt. max_new_tokens=20: Limit the output to 20 new tokens.
```

```python
# 8. Decode and Print Output
# Purpose: Convert the generated token IDs back into readable text and print the result.

print(tokenizer.decode(generation_output[0]))
```

## Comparing Trained LLM Tokenizers

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

colors_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence, tokenizer_name):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids
    for idx, t in enumerate(token_ids):
        print(
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m' +
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' '
        )
```

```python
text = """
English and CAPITALIZATION
🎵 鸟
show_tokens False None elif == >= else: two tabs:"    " Three tabs: "       "
12.0*50=600
"""
```

```python
show_tokens(text, "bert-base-uncased")
```

```python
show_tokens(text, "bert-base-cased")
```

```python
show_tokens(text, "gpt2")
```

## Recommending songs by embeddings

```python
import pandas as pd
from urllib import request

# Get the playlist dataset file
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')

# Parse the playlist dataset file. Skip the first two lines as
# they only contain metadata
lines = data.read().decode("utf-8").split('\n')[2:]

# Remove playlists with only one song
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]

# Load song metadata
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')
songs_file = songs_file.read().decode("utf-8").split('\n')
songs = [s.rstrip().split('\t') for s in songs_file]
songs_df = pd.DataFrame(data=songs, columns = ['id', 'title', 'artist'])
songs_df = songs_df.set_index('id')
```

```python
 print( 'Playlist #1:\n ', playlists[0], '\n')
```

```python
print( 'Playlist #2:\n ', playlists[1])
```

```python
from gensim.models import Word2Vec

# Train our Word2Vec model , Paper link : https://arxiv.org/pdf/1301.3781
model = Word2Vec(
    playlists, vector_size=32, window=20, negative=50, min_count=1, workers=4
)
```

```python
song_id = 2172

# Ask the model for songs similar to song #2172
model.wv.most_similar(positive=str(song_id))
```

```python
print(songs_df.iloc[2172])
```

```python
import numpy as np

def print_recommendations(song_id):
    similar_songs = np.array(
        model.wv.most_similar(positive=str(song_id),topn=10)
    )[:,0]
    return  songs_df.iloc[similar_songs]

# Extract recommendations
print_recommendations(2172)
```
